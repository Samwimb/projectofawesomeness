```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as stats
import pycountry
```

```python
hdi_data = pd.read_csv("../Resources/Human_Development_Index.csv")
```

```python
hdi_data.head(2)
```

```python
# Rename multiple columns in one go with a larger dictionary
hdi_data.rename(
    columns={
        "2016": "HDI 2016",
    },
    inplace=True)
```

```python
hdi_data.head(2)
```

```python
#Removing blank spaces from Country Name
hdi_data['Country'] = hdi_data['Country'].str.strip()
```

```python
for i, row in hdi_data.iterrows():
  try:
      hdi_data.loc[i, 'country code'] = pycountry.countries.lookup(row['Country']).alpha_3
  except LookupError:
       if hdi_data.loc[i, 'Country'] == 'Bolivia (Plurinational State of)':
           hdi_data.loc[i, 'country code'] = 'BOL'
       elif hdi_data.loc[i, 'Country'] == 'Congo (Democratic Republic of the)':
           hdi_data.loc[i, 'country code'] = 'COD'
       elif hdi_data.loc[i, 'Country'] == 'Eswatini (Kingdom of)':
           hdi_data.loc[i, 'country code'] = 'SWZ'
       elif hdi_data.loc[i, 'Country'] == 'Iran (Islamic Republic of)':
           hdi_data.loc[i, 'country code'] = 'IRN'
       elif hdi_data.loc[i, 'Country'] == 'Korea (Republic of)':
           hdi_data.loc[i, 'country code'] = 'KOR'
       elif hdi_data.loc[i, 'Country'] == 'Micronesia (Federated States of)':
           hdi_data.loc[i, 'country code'] = 'FSM'
       elif hdi_data.loc[i, 'Country'] == 'Moldova (Republic of)':
           hdi_data.loc[i, 'country code'] = 'MDA'
       elif hdi_data.loc[i, 'Country'] == 'Hong Kong, China (SAR)':
           hdi_data.loc[i, 'country code'] = 'HKG'
       elif hdi_data.loc[i, 'Country'] == 'Tanzania (United Republic of)':
           hdi_data.loc[i, 'country code'] = 'TZA'
       elif hdi_data.loc[i, 'Country'] == 'Venezuela (Bolivarian Republic of)':
           hdi_data.loc[i, 'country code'] = 'VEN'
       else:
           hdi_data.loc[i, 'country code'] = '*unk'

hdi_data.to_csv("../Resources/HDI_Data_cleaned.csv", index=False)
```
